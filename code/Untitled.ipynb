### Examining distributions

```python
def plot_build(column):
    plt.figure(figsize=(10, 6))
    sns.histplot(train[column], kde=True, bins=20, color='blue')
    plt.title(f'Distribution of column')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()

    # Boxplot
    plt.figure(figsize=(8, 5))
    sns.boxplot(x=train[column], color='green')
    plt.title(f'Boxplot of {column}')
    plt.xlabel(column)
    plt.show()
```
